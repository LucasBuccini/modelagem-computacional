```python
from matplotlib.pyplot import plot, grid, legend, xlabel, ylabel
from numpy import array, gradient, sin, sqrt, linspace

def f(x):
    y = sin(0.5 * sqrt(x)) / x
    return y

xl = linspace(1,7,100)
x = array([1, 3, 5, 7])
x2 = array([1, 2, 3, 4, 5, 6, 7])

# cálculo da derivada usando x (passo maior)
derivada = gradient(f(x), x)

# cálculo da derivada usando x2 (passo menor)
derivada2 = gradient(f(x2), x2)

plot(xl, f(xl), label='f(x)')
plot(x, derivada, 'rx', label='dfdx com dx=2')
plot(x2, derivada2, 'go', label='dfdx com dx=1')

xlabel('x')
ylabel('f(x) e dfdx')
legend()
grid()
```
