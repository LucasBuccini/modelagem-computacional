```python
from scipy.optimize import *
from matplotlib.pyplot import *
from numpy import linspace
f = lambda c: (2*c)/(4+0.8*c+c**2+0.2*c**3)

fx = lambda c: -1*((2*c)/(4+0.8*c+c**2+0.2*c**3))

x = linspace(0,10,100)

plot(x,f(x),label="função")
legend()
grid()

valor = minimize(fx,0)
print(valor.x)

y = valor.x
plot(y,f(y),'ro',label="maximo")
legend()
```

```python
from numpy import dot

A = array([[0.52,0.20,0.25],[0.30,0.50,0.20],[0.18,0.30,0.55]])
b = array([[4800],[5810],[5690]])

I = inv(A)

resp = dot(I,b)

print(resp)
```
