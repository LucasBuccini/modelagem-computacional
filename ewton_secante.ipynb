```python
from matplotlib.pyplot import *
from numpy import *
from scipy.optimize import newton

f = lambda x: x**3 - 6*x**2+11*x - 6.1
fl = lambda x: 3*x**2 - 12*x + 11

x = linspace(0.5,3.5,100)
plot(x,f(x),label="Função")
grid()
legend()
```

```python
#newton
plot(x,f(x),label="Função")
x0= 3.5
x1 = x0 - f(x0)/fl(x0)
print(f"X = {x1} na 1° interação")

for i in range(2):
    x0=x1
    x1 = x0 - f(x0)/fl(x0)
    print(f"X = {x1} na {i+2}° interação ")
    plot(x1,f(x1),"bd",label="Valor de newton")

plot(x1,f(x1),"rd",label="newton")
grid()
legend()
print(x1)
print(f"Erro de {f(x1)}")
```

```python
plot(x,f(x),label="Função")

x0 = 0.5
x1 = 3.5
x2 = x1 - (f(x1)*(x0-x1))/(f(x0)-f(x1))

print(f"X = {x2} na 1° interação")

for i in range(2):
    x0=x1
    x1 = x2
    x2 = x1 - (f(x1)*(x0-x1))/(f(x0)-f(x1))
    print(f"X = {x2} na {i+2}° interação ")

plot(x2,f(x2),"rd",label="Secante")
grid()
legend()
print(x2)
print(f"Erro de {f(x2)}")
    
```
