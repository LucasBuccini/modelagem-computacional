```python
from numpy import linspace
from math import e
import matplotlib.pyplot as plt
from scipy.optimize import bisect
```

```python
P_U_MAX = 75000
K_U = 0.045
P_U_MIN = 100000
P_S_MAX = 300000
P_0 = 10000
K_S = 0.08
```

```python
pu = lambda t : P_U_MAX * e ** (-K_U*t) + P_U_MIN
ps = lambda t: (P_S_MAX / (1+ (P_S_MAX / (P_0**-1)) * (e** (-K_S*t)))) * 1.2

root = lambda t: pu(t) - ps(t)
```

```python
dominio = linspace(200,520,100)

xl = float(input("Qual o Xl: "))
xu = float(input("Qual o xu: "))

if root(xu)*root(xl) < 0:
    raiz = bisect(root,xl,xu,full_output=True)


    plt.plot(dominio,(pu(dominio)),label = "pu")
    plt.plot(dominio,(ps(dominio)),label = "ps")
    plt.plot(dominio,root(dominio),label="raiz")
    plt.plot(raiz[0],root(raiz[0]),"ro")
    plt.legend()
    plt.grid()
    print(raiz)
    print(f'Ps vai ser 20% maior que Pu quando t for igual a = {raiz[0]:.4f}')
else:
    print("A função nos chutes iniciais devem ter sinais opostos")
```
