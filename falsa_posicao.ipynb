```python
from matplotlib.pyplot import * 
from numpy import *
fx = lambda x: x**10 - 1

x = linspace(0,1.4,100)
plot(x,fx(x),"b-",label="Funo")
xl = float(input("Qual o x inicial: "))
xu = float(input("Qual o x final: "))
erro = float(input("Qual a tolerancia do erro: "))
xr = xu - ((fx(xu)*(xl-xu))/(fx(xl)-fx(xu)))
print(f"xl = {xl}, xu = {xu}, xr = {xr}, erro = {fx(xr)}")
cont=0
if(fx(xl)*fx(xu)<0):
    while(abs(fx(xr))>erro):
        if (fx(xl)*fx(xr)<0):
            xu = xr
        elif (fx(xr)*fx(xu)<0):
            xl=xr
        
        print(f"xl = {xl}, xu = {xu}, xr = {xr}, erro = {fx(xr)}")
        xr = xu - ((fx(xu)*(xl-xu))/(fx(xl)-fx(xu)))
        cont+=1
    print(f"Raiz encontrada em x = {xr}, com {cont} interaes")
    plot(xr,0,"ro",label="Raiz")
    grid()
    legend()
else:
    print("Valores invalidos, intervalo fora da raiz")

```

```python
f = lambda x: -4 + 1.8*x**2 + 1.2*x**3 - 0.3*x**4
ff = lambda x: 3.6*x + 3.6*x**2 - 1.2*x**3
fff = lambda x: 3.6 + 7.2*x - 3.6*x**2

x = linspace(0,5,100)
plot(x,f(x),label="Funo")
grid()
xl = float(input("Qual o x inicial: "))
xu = float(input("Qual o x final: "))
erro = float(input("Qual a tolerancia do erro: "))
xr = xu - ((ff(xu)*(xl-xu))/(ff(xl)-ff(xu)))
print(f"xl = {xl}, xu = {xu}, xr = {xr}, erro = {ff(xr)}")

if(ff(xl)*ff(xu)<=0):
    while(abs(ff(xr))>erro):
        if (ff(xl)*ff(xr)<0):
            xu = xr
        elif (ff(xr)*ff(xu)<0):
            xl=xr
        
        print(f"xl = {xl}, xu = {xu}, xr = {xr}, erro = {fx(xr)}")
        xr = xu - ((ff(xu)*(xl-xu))/(ff(xl)-ff(xu)))
        cont+=1
    if(fff(xr)<0):
        print(f"Maximo encontrad0 em x = {xr}, com {cont} interaes")
        plot(xr,f(xr),"ro",label="Raiz")
        legend()
    else:
        print(f"Intervalo possui um minimo em {xr}, nao foi possivel encontrar o maximo, digite um intervalo que no passe por esse ponto")
else:
    print("Valores invalidos, intervalo fora da raiz")
```
