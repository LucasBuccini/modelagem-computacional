```python
from scipy.interpolate import lagrange
from numpy import array, poly1d, polyfit, linspace, sin
from matplotlib.pyplot import plot, xlabel, ylabel, legend, title, grid

f = lambda x: 1/(1+25*x**2)

dominio = linspace(-1,1,100)
x= linspace(-1,1,21)

plot(dominio,f(dominio),label="Função")


fx = f(x)
b = polyfit(x,fx,6)
c = poly1d(b)
plot(dominio,c(dominio),label="Polinomio encontrado")
plot(x, fx, 'ro', markersize=5, label="Pontos tabelados")

grid()
legend()
```

```python
from math import pi
t = lambda t: sin(t)**2
dominio = linspace(0,2*pi,100)
x = linspace(0,2*pi,8)
tx = t(x)
f = lagrange(x,tx)

plot(dominio,t(dominio),label="Função")
plot(dominio,f(dominio),label="Lagrange")

legend()
grid()

```

```python
from scipy.integrate import trapezoid
from numpy import e
from matplotlib.pyplot import fill_between
f = lambda x: 9.8*68.1/12.5*(1-e**(-(12.5/68.1)*x))
x = linspace(0,10,100)
y= linspace(0,10,10)
fx=f(x)
fy=f(y)
plot(x,f(x),label="Função")
plot(y,f(y),"rd-",label="Pontos")
fill_between(y,fy,color="lightgreen",alpha=0.5)
plot
area = trapezoid(fy, y)
exato = 289.43515
legend()
grid()
print(f"Area encontrada= {area}")
print(f"Erro de {(exato-area)/exato*100:.2f}%")
```

```python
from scipy.integrate import simpson

f = lambda x: 0.2 + 25*x-200*x**2 +675*x**3 - 900*x**4 + 400*x**5
x = linspace(0,0.8,5)
dominio = linspace(0,0.8,100)
dominio1 = linspace(0,0.4,100)
plot(dominio,f(dominio),label="Função")

simp = simpson(f(x),x)
fx=f(x)
print(simp)

b1 = polyfit(x,f(x),4)
c1 = poly1d(b1)
plot(x,f(x),"rd",label="Pontos")
plot(dominio,c1(dominio))
fill_between(dominio,c1(dominio),color="lightgreen",alpha=0.5)
legend()
grid()
```

```python
from scipy.integrate import solve_ivp

f = lambda x,y: 4*e**(0.8*x) - 0.5*y

xinicial = 0
xfinal = 4
yinicial = 2
valores = linspace(0,4,100)
sol = solve_ivp(f,[xinicial,xfinal],[yinicial],t_eval=valores)
plot(valores,sol.y[0,:],label="Runge Kutta")

grid()
legend()

print(len(sol.y[0]))
```
