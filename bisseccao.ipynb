```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as sp

f = lambda x: 7*(2-0.9**x) -10
xy = np.linspace(4,8,100)
plt.plot(xy,f(xy),'r-',label="função")
plt.grid()


xl = 4.0
xu = 8.0
raiz = sp.bisect(f,xl,xu)
plt.plot(raiz,f(raiz),"bd",label="Solução")
plt.legend()
print(raiz)
```
