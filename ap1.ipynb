```python
from numpy import linspace
from math import e
from matplotlib.pyplot import * 
from scipy.optimize import bisect
```

```python
P_U_MAX = 75000
K_U = 0.045
P_U_MIN = 100000
P_S_MAX = 300000
P_0 = 10000
K_S = 0.08
```

```python
pu = lambda t : P_U_MAX * e ** (-K_U*t) + P_U_MIN
ps = lambda t: (P_S_MAX / (1+ (P_S_MAX / (P_0**-1)) * (e** (-K_S*t)))) * 1.2

root = lambda t: pu(t) - ps(t)
```

```python
x = linspace(200,520,100)

xl = float(input("Qual o x inicial: "))
xu = float(input("Qual o x final: "))

if (root(xu)*root(xl)<0):
    raiz = bisect(root,xl,xu,full_output=True)
    plot(x,ps(x),label="ps")
    plot(x,pu(x),label="pu")
    plot(x,root(x),label="root")
    grid()
    legend()
else:
    print("Intervalo invalido")
plot(raiz[0],root(raiz[0]),"ro",label="Raiz")
print(raiz)
```
